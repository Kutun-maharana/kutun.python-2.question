```python
#30.Dictionary with student names and marks. Display students who scored more than 80.

students = {
    "Siva": 85,
    "Ravi": 70,
    "Anu": 90,
    "Kiran": 75,
    "John": 95
}

for name, marks in students.items():
    if marks > 80:
        print(name, marks)

```
