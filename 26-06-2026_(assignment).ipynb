# -:  **INSTANCE METHOD** :-



**Definition:** An instance method is a method that works with a specific object of a class. It can access and modify the data of that object . **Real-world example:**

 **A Student class:** each student has their own name and marks. A method to display a particular student's details is an instance method.

**EXAMPLES OF INSTANCE METHOD:-**

```python
class Student:                                 # STUDENT DETAILS
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(self.name, self.marks)

s1 = Student("Rahul", 90)
s1.display()
```

```python
class BankAccount:                                 # BANK ACCOUNT
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def show_balance(self):
        print("Balance:", self.balance)

account = BankAccount("Amit", 5000)
account.show_balance()
```

```python
class Car:                                        # CAR INFORMATION
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def car_info(self):
        print(self.brand, self.model)

car1 = Car("Toyota", "Innova")
car1.car_info()
```

```python
class Employee:                                  # EMPLOYEE DETAILS
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_employee(self):
        print(self.name, self.salary)

emp1 = Employee("Priya", 40000)
emp1.show_employee()
```

```python
class Mobile:                                    # MOBILE PHONE
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def show_mobile(self):
        print(self.brand, self.price)

phone = Mobile("Samsung", 20000)
phone.show_mobile()
```

```python
class Rectangle:                                # RECTANGLE AREA
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        area = self.length * self.width
        print("Area:", area)

rect = Rectangle(10, 5)
rect.calculate_area()
```

# -: **CLASS METHOD** :-

**Definition:**
A class method is a method that works with the class itself instead of a specific object. It is used to access or modify class-level information.

**Real-world example:**

**A Company class:** if we want to store the total number of employees in the company, a class method can manage that shared information.

**EXAMPLES OF CLASS METHOD:-**

```python
class Student:                        # Counting Students
    total_students = 0

    def __init__(self, name):
        self.name = name
        Student.total_students += 1

    @classmethod
    def get_total_students(cls):
        return cls.total_students

s1 = Student("Akash")
s2 = Student("Priya")

print(Student.get_total_students())
```

```python
class Student:                       # Counting Number of Objects Created
    count = 0

    def __init__(self, name):
        self.name = name
        Student.count += 1

    @classmethod
    def total_students(cls):
        return cls.count

s1 = Student("Rahul")
s2 = Student("Priya")

print(Student.total_students())
```

```python
class Employee:                      # Changing Company Name
    company = "ABC Ltd"

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

Employee.change_company("XYZ Ltd")
print(Employee.company)
```

```python
class Employee:                      # Employee company details
    company = "TCS"

    @classmethod
    def show_company(cls):
        print(cls.company)

Employee.show_company()
```

```python
class Bank:                          # Bank interest rate
    interest_rate = 5

    @classmethod
    def change_rate(cls, rate):
        cls.interest_rate = rate

Bank.change_rate(7)

print(Bank.interest_rate)
```

```python
class Product:                        # Updating product discount
    discount = 10

    @classmethod
    def update_discount(cls, value):
        cls.discount = value

Product.update_discount(20)

print(Product.discount)
```

# **-: STATIC METHOD :-**

**Definition:**
A static method is a method that does not depend on object or class data. It performs a general task related to the class.

**Real-world example:**

**A Calculator class:** adding two numbers is a general operation and does not need employee/student information, so it can be a static method.

**EXAMPLES OF STATIC METHOD:-**

```python
class Calculator:                  # Calculator (Addition)

    @staticmethod
    def add(a, b):
        return a + b

print(Calculator.add(5, 3))
```

```python
class Number:                       # Check Even or Odd

    @staticmethod
    def check_even(num):
        if num % 2 == 0:
            return "Even"
        else:
            return "Odd"

print(Number.check_even(10))
```

```python
class Temperature:                   # Temperature Conversion

    @staticmethod
    def celsius_to_fahrenheit(c):
        return (c * 9/5) + 32

print(Temperature.celsius_to_fahrenheit(25))
```

```python
class Rectangle:                     # Calculate Area of Rectangle

    @staticmethod
    def area(length, width):
        return length * width

print(Rectangle.area(5, 4))
```

```python
class User:                           # Password Validation

    @staticmethod
    def check_password(password):
        if len(password) >= 8:
            return "Valid Password"
        else:
            return "Invalid Password"

print(User.check_password("python123"))
```

```python
class Number:                        # Find Maximum Number

    @staticmethod
    def maximum(a, b):
        return max(a, b)

print(Number.maximum(10, 20))
```

  -: **DIFFERENCE BETWEEN INSTANCE, CLASS, STATIC METHOD** :-

| Feature         | Instance Method                   | Class Method                 | Static Method                          |
| --------------- | --------------------------------- | ---------------------------- | -------------------------------------- |
| Works with      | Object (instance)                 | Class                        | Neither object nor class               |
| First parameter | `self`                            | `cls`                        | No required parameter                  |
| Accesses        | Instance variables                | Class variables              | Only local data                        |
| Used for        | Object-specific operations        | Class-level operations       | General utility operations             |
| Example         | Student details, Employee details | Total students, company name | Calculator operations, data validation |
