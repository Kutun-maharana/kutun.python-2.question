# SINGLE-LEVEL INHERITANCE



```python
class Person:    # Example 1: Person → Student
    def show(self):
        print("I am a person")

class Student(Person):
    def study(self):
        print("I am studying")

s = Student()
s.show()
s.study()
```

```python
class Animal:        # Example 2: Animal → Dog
    def eat(self):
        print("Animal eats food")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.eat()
d.bark()
```

```python
class Vehicle:          # Example 3: Vehicle → Car
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")

c = Car()
c.start()
c.drive()
```

# MULTILEVEL INHERITANCE

```python
class Person:         # Example 1: Person → Student → CollegeStudent
    def name(self):
        print("I am a person")

class Student(Person):
    def study(self):
        print("I study")

class CollegeStudent(Student):
    def exam(self):
        print("I give exams")

c = CollegeStudent()
c.name()
c.study()
c.exam()
```

```python
class Animal:            # Example 2: Animal → Dog → Puppy
    def eat(self):
        print("Animal eats")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Puppy(Dog):
    def play(self):
        print("Puppy plays")

p = Puppy()
p.eat()
p.bark()
p.play()
```

```python
class Device:           # Example 3: Device → Mobile → Smartphone
    def power(self):
        print("Device has power")

class Mobile(Device):
    def call(self):
        print("Mobile can call")

class Smartphone(Mobile):
    def internet(self):
        print("Smartphone uses internet")

s = Smartphone()
s.power()
s.call()
s.internet()
```
